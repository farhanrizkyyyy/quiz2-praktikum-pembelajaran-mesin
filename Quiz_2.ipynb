```python
df['Unnamed: 0'] = df['Unnamed: 0'].fillna(0)
```

```python
df.describe(include='all')
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

correlation = df.corr()
plt.figure(figsize=(15, 10))
sns.heatmap(correlation[(correlation >= 0.5) | (correlation <= -0.4)], cmap='viridis', vmax=1.0, vmin=-1.0, linewidths=0.1, 
            annot=True, annot_kws={"size": 8}, square=True).set_title('Correlation')
plt.show()
```

```python
desc_total = df.groupby('description')['Unnamed: 0'].sum().sort_values()
desc_total.plot(kind='barh', fontsize=10, figsize=(8, 40), title='Total Description')
plt.show()

title_total = df.groupby('Title')['Unnamed: 0'].sum().sort_values()
title_total.plot(kind='barh', fontsize=10, figsize=(8, 40), title='Total Title')
plt.show()
```
